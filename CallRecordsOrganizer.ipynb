```python
import shutil
import os
#from pathlib import Path
import re
```

```python
source = "D:/Dell/Desktop/CallRecordingPapa/Call"
```

```python
for file in os.listdir(source):
    file_name = file.strip()
    folder_name = ''.join(re.findall('^Call recording (.+?)_',file_name))
    folder_path = f'D:/Dell/Desktop/CallRecordingPapa/{folder_name}'
    os.makedirs(folder_path, exist_ok = True)
    source_path = os.path.join(source,file)
    destination_path = os.path.join(folder_path,file)
    shutil.move(source_path,destination_path)
    print(f'Moved: {file:70} to Folder: {folder_name}')

```
